```python
# Dependencies 
import pandas as pd
import json
import numpy as np
```

```python
# load CSV
purchase_data1 = open('HeroesOfPymoli/purchase_data.JSON').read()
purchase_data1
```

```python
purchase_pd1 = pd.read_json('HeroesOfPymoli/purchase_data.JSON', orient = 'columns')
purchase_df1 = pd.DataFrame(purchase_pd1)
purchase_df1.head(80)
```

```python
purchase_pd2 = pd.read_json('HeroesOfPymoli/purchase_data2.JSON', orient = 'columns')
purchase_df2 = pd.DataFrame(purchase_pd2)
purchase_df2.head(5)
```

```python
combined_data = []
combined_data = pd.concat([purchase_df1, purchase_df2])
#orient :{'columns'} 
combined_data.head()
```

```python
#Player Count
total_players = len(combined_data)
print('Total Number of Players')
total_players
```

```python
print('Purchasing Analysis')

#Number of Unique Items
unique_item = len(pd.unique(combined_data['Item Name']))
print('Number of Unique Items: ', unique_item)
```

```python
#Average Purchase Price

average_purchase_price = combined_data['Price'].mean()
#round(average_purchase_price,2)

print('Average Purchase Price: ',average_purchase_price)
```

```python
#Total Number of Purchases
total_purchase = len(combined_data['Item ID'])
print('Total Number of Purchases: ',total_purchase)
```

```python
#Total Revenue
total_revenue = combined_data['Price'].sum()
print('Total Revenue: ',total_revenue )
```

```python
print('Gender Demographics')
#Percentage and Count of Male Players
total_gender = combined_data["Gender"].count()
total_gender
male = combined_data["Gender"].value_counts()['Male']
male
female = combined_data["Gender"].value_counts()['Female']
female
non_gender_specific = total_gender - male - female
print(f" Total: {total_gender}\n Male: {male}\n Female: {female}\n non_specfic: {non_gender_specific}")
#male_percent
male_percent = male/total_gender

print("Male_percent:""{:.2%}".format(male_percent))
female_percent = female/total_gender
print("Female_percent:""{:.2%}".format(female_percent))
non_specific_percent =  non_gender_specific/total_gender
print("non_specific_percent:""{:.2%}".format(non_specific_percent))
```

```python
#Average Purchase Price
male_filtered = combined_data[combined_data.Gender=="Male"]
avg_male_purchase_price = male_filtered["Price"].mean()
avg_male_purchase_price
print("avg_male_purchase_price:""${:.3}".format(avg_male_purchase_price))
female_filtered = combined_data[combined_data.Gender=="Female"]
avg_female_purchase_price = female_filtered["Price"].mean()
avg_female_purchase_price
print("avg_female_purchase_price:" "${:.3}".format(avg_female_purchase_price))
```

```python
#Total Purchase Value
male_purchase = combined_data["Gender"].value_counts()['Male']
female_purchase = combined_data["Gender"].value_counts()['Female']
male_purchase_value = male_filtered["Price"].sum()
print("${:,.2f}".format(male_purchase_value))
female_purchase = combined_data[combined_data.Gender=="Female"]
female_purchase_value = female_purchase["Price"].sum()
female_purchase_value
print("${:,.2f}".format(female_purchase_value))
```

```python
#Normalized Totals

male_max_price = male_filtered["Price"].max()
male_min_price = male_filtered["Price"].min()

normalized_male_price = (male_filtered["Price"]-male_min_price)/(male_max_price-male_min_price)
print("Normalize Male Purchase Total: ", "${:,.2f}".format(normalized_male_price.sum()))

normalized_female_price = (female_filtered["Price"]-male_min_price)/(male_max_price-male_min_price)
print("Normalize Female Purchase Total: ", "${:,.2f}".format(normalized_female_price.sum()))
```

```python
#Top Spenders
#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

top_spenders = combined_data.groupby('SN')['Price'].agg(['count','mean', 'sum'])
#top_spenders.rename(columns={"SN": "SN", "count": "C1", "mean": "M1", "sum"" "S1""})
#top_spenders = combined_data.sort_values(['Price'], ascending=False).groupby("SN", as_index=False)["Price"].sum().head(5)
#top_spenders.reset_index(inplace=True)
##top_spenders.sort_values(by=['Price'], ascending =False)
print(top_spenders[:5])
```

```python
#Most Popular Items

#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value

top_popular = combined_data.groupby(['Item ID', 'Item Name'])['Price'].agg(['count', 'sum'])
#top_popular = combined_data.groupby(['Item ID','Item Name']).agg(['Item ID':[count], 'Price':sum])
top_popular.reset_index(inplace=True)
top_popular.sort_values(by='count' , ascending=False)
print(top_popular[:5])
```

```python
#Most Profitable Items

#Identify the 5 most profitable items by total purchase value, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
most_profitable = combined_data.groupby(['Item ID', 'Item Name'])['Price'].agg(['count', 'sum'])
#top_popular = combined_data.groupby(['Item ID','Item Name']).agg(['Item ID':[count], 'Price':sum])
most_profitable.reset_index(inplace=True)
most_profitable.sort_values(by='sum' , ascending=False)
print(most_profitable[:5])
```
